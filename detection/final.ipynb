```python
#drawVertices
import io
from PIL import Image, ImageDraw, ImageFont

def drawVertices(filename, image_source, vertices, display_text=''):
    pillow_img = Image.open(io.BytesIO(image_source))

    draw = ImageDraw.Draw(pillow_img)
    for i in range(len(vertices) - 1):
        draw.line(((vertices[i].x, vertices[i].y), (vertices[i + 1].x, vertices[i + 1].y)),
                fill='green',
                width=8
        )

    draw.line(((vertices[len(vertices) - 1].x, vertices[len(vertices) - 1].y),
               (vertices[0].x, vertices[0].y)),
               fill='green',
               width=8
    )

    font = ImageFont.truetype('arial.ttf', 16)
    draw.text((vertices[0].x + 10, vertices[0].y),
              font=font, text=display_text, 
              fill=(255, 255, 255))
#     pillow_img.show()
    pillow_img.save("zxc/frame_to_video/"+filename.split('.')[0].split('/')[-1] + ".jpg")
```

```python
import os, io
# from draw_vertice import drawVertices
from google.cloud import vision
import pandas as pd
def google_api(filename):
    client = vision.ImageAnnotatorClient()

    with io.open(filename, 'rb') as image_file:
        content = image_file.read()

    image = vision.types.Image(content=content)
    response = client.logo_detection(image=image)
    logos = response.logo_annotations

    for logo in logos:
    #     print('Logo Description:', logo.description)
    #     print('Confidence Score:', logo.score)
    #     print('-'*50)
        vertices = logo.bounding_poly.vertices
    #     print('Vertices Values {0}'.format(vertices))
        drawVertices(filename, content, vertices, logo.description)
        break
```

```python
#프레임 합치기
import cv2
import numpy as np
import os
from os.path import isfile, join

import re
def solution(files):
    file_lst = [re.split('([0-9]+)',i) for i in files]
    file_lst.sort( key = lambda x : ( x[0].lower(), int(x[1]) ) )
    return [''.join(lst) for lst in file_lst]

pathIn = 'zxc/nike_movie2'
fps = 24
frame_array = []
files = [f for f in os.listdir(pathIn) if isfile(join(pathIn, f))]
files = solution(files)
files_short = files[:300]
files=files[500:]
for i in files:
    print (i)
    google_api("zxc/nike_movie2/"+ i)
    
# pathIn = 'zxc/frame_to_video'
# pathOut = 'zxc/nike.avi'

# for i in range(len(files_short)):
#     filename=pathIn + '/' + files[i].split('.')[0].split('/')[-1] + ".jpg"
#     #reading each files
#     img = cv2.imread(filename)
#     height, width, layers = img.shape
#     size = (width,height)
    
#     #inserting the frames into an image array
#     frame_array.append(img)
# out = cv2.VideoWriter(pathOut,cv2.VideoWriter_fourcc(*'DIVX'), fps, size)
# for i in range(len(frame_array)):
#     # writing to a image array
#     out.write(frame_array[i])
# out.release()
```

```python
import time

pathIn = 'zxc/frame_to_video/'
pathOut = 'zxc/nike.avi'

files_2 = [f for f in os.listdir(pathIn) if isfile(join(pathIn, f))]
files_2 = solution(files_2)

for i in files_2:
    print(i)
    filename=pathIn + '/' + i.split('.')[0].split('/')[-1] + ".jpg"
    #reading each files
    img = cv2.imread(filename)
    height, width, layers = img.shape
    size = (width,height)
    
    #inserting the frames into an image array
    frame_array.append(img)
out = cv2.VideoWriter(pathOut,cv2.VideoWriter_fourcc(*'DIVX'), fps, size)
for i in range(len(frame_array)):
    # writing to a image array
    out.write(frame_array[i])
out.release()
```
